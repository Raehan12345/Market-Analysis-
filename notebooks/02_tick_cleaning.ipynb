```python
# import libraries and custom cleaner module
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os

# setup path for src imports
sys.path.append(os.path.abspath('../'))
from src.cleaner import clean_tick_data
```

```python
# load the raw data
input_path = '../data/raw/spy_raw_1m.csv'
df_raw = pd.read_csv(input_path)

# check the initial length
print(f"original row count: {len(df_raw)}")
```

```python
# clean the data using a z-score threshold of 3.0
df_clean = clean_tick_data(df_raw, price_col='Close', z_threshold=3.0)

# check how many outlier rows were removed
print(f"cleaned row count: {len(df_clean)}")
print(f"removed {len(df_raw) - len(df_clean)} outliers or empty rows")
```

```python
# save the processed data to the processed folder
output_path = '../data/processed/spy_cleaned_1m.csv'
df_clean.to_csv(output_path, index=False)
print("cleaned data saved and ready for analysis.")
```
